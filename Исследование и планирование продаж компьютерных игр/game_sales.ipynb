По информации из википедии данный рейтинг присваивает Entertainment Software Rating Board (ESRB) — негосударственная организация, основное направление деятельности — принятие и определение рейтингов для компьютерных видеоигр и другого развлекательного программного обеспечения в США и Канаде. ESRB была основана в 1994 году ассоциацией Interactive Digital Software Association (современное название — «Entertainment Software Association»). К началу 2003 года рейтинги ESRB получили более 8000 продуктов от 350 издателей.  
Рейтинг игры напечатан на упаковке, содержится в её рекламе и указан на сайте игры. Компании не обязаны предоставлять игру на определение рейтинга до начала официальных продаж, так как её содержимое может значительно измениться в ходе разработки.  
Согласно системе ESRB приняты следующие обозначения:
- «EC» («Early childhood») — «Для детей младшего возраста»
- «E» («Everyone») — «Для всех» Первоначально "K-A" ("Kids to Adults")
- «E10+» («Everyone 10 and older») — «Для всех от 10 лет и старше»
- «T» («Teen») — «Подросткам»
- «M» («Mature») — «Для взрослых»
- «AO» («Adults Only 18+») — «Только для взрослых»
- «RP» («Rating Pending») — «Рейтинг ожидается»
- «RP 17+» («Rating Pending») — «Рейтинг ожидается - примерно 17+»

Учитывая тот факт, что рейтинг применим к контенту в США и Канаде, можно предположить, что в этом и есть причина пропусков. Заменим пропущенные значение на неизвестные, чтобы не исключить их при дальнейшем анализе влияния рейтинга ESRB на продажи в отдельном регионе.

```python
# Заполняем пропуски на значение unknown

data['rating'].fillna('unknown', inplace = True)
```

```python
# Также заменим устаревшее обозначение рейтинга K-A на E

data['rating'].replace('K-A', 'E', inplace = True)
```

#### Исследование игр по жанрам

Для анализа общих продаж игр по жанрам построим комбинированную диаграмму, где столбчатая диаграмма будет отражать суммарные продажи по каждому жанру, а линейный график показывать общее кол-во игр, выпускаемых для разных платформ в каждом жанре.

```python
# Создадим сводную таблицу общих продаж и выпускаемых игр по жанрам

genre_sales = (new_data.query('platform in @top_platform')
               .pivot_table(index='genre', values='total_sales', aggfunc=['sum', 'count']).reset_index()) 
genre_sales.columns = ['Жанры', 'Сумма продаж', 'Количество игр']
genre_sales.sort_values(by='Сумма продаж', ascending=False)
```

```python
# Строим комбинированную диаграмму

fig, ax1 = plt.subplots(figsize=(19,6))
color = 'tab:green'
ax1.set_title('Общие продажи игр по жанрам', fontsize=16)
ax1.set_xlabel('Жанры', fontsize=16, color=color)
ax1.set_ylabel('Сумма продаж', fontsize=16)
ax2 = sns.barplot(x='Жанры', y='Сумма продаж', data = genre_sales, palette='summer')
ax1.tick_params(axis='y')
ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Количество игр', fontsize=16, color=color)
ax2 = sns.lineplot(x='Жанры', y='Количество игр', data = genre_sales, sort=False, color=color)
ax2.tick_params(axis='y', color=color)
plt.show()
```

Как видно из графика, наибольшие продажи приносит жанр Action (экшен) - 160 млн.копий, он же является рекордсменом по кол-ву игр, выпускаемых для различных платформ - 380 релизов, с небольшим отрывом идёт жанр Shooter (стрелялки) с результатом 145 млн. и почти 100 игр, за почётное 3-ье место борятся 2 жанра: Role-Playing (ролевые игры)  - 85 млн. и 130 игр и Sports (спорт)	75млн. и 95 игр, т.е. при почти равном кол-ве игр с Shooter, Sports приносит почти в 2 раза меньшие продажи. Остальные жанры сильно проигрывают ТОП-3: на Fighting (единоборства), Misc (смешанный жанр), Racing (гонки) в среднем пришлось по 24 млн.копий, хотя разброс по кол-ву игр, выпускаемых в данных жанрах больше, чем в 2 раза - для Fighting - 30 игр, а для Misc - 70, Racing занимает золотую середину. Отстаёт от них в среднем на 7 млн. жанр Platform (платформер) на который приходится 30 игр. С результатом 11 и 12 млн.копий идут жанры Simulation (симулятор) и Adventure (приключения, квест), причём, в жанре Adventure выпущено 75 игр, а Simulation в 2 раза меньше. И наименьшую долю продаж имеют жанры Strategy (стратегия) и Puzzle (головоломка). Puzzle выпущено всего 10 игр, Strategy в 3 раза больше. 

Теперь, чтобы определить жанры с наибольшим потенциалом и сравнить статистики для их выборок, воспользуемся методом describe() и графиком "ящик с усами", которые также позволят нам свести продажи каждой игры, выпущенной для разных платформ.

```python
# Посмотрим статистики для всех жанров

(new_data.query('platform in @top_platform')
 .pivot_table(index='name', columns='genre', values='total_sales', aggfunc='sum')
 .describe(percentiles=[0.005, 0.25, 0.5, 0.75, 0.95]).round(2))
```

```python
# Создадим переменную со сводными данными продаж каждой игры, выпущенной для разных платформ, по всем жанрам

genre_median = (new_data.query('platform in @top_platform')
                .pivot_table(index=['genre', 'name'], values='total_sales', aggfunc='sum').reset_index())
```

### Проверка гипотез

Поскольку H₀ всегда формулируется так, чтобы использовать знак равенства, чтобы все данные лежали в районе изначально предполагаемого значения, сформулируем нулевую гипотезу и альтернативную гипотезу, противоречащую нулевой.
Поскольку выполняются необходимые условия для применение t-теста: генеральные совокупности не зависят друг от друга и благодаря достаточности размеров выборки выборочные средние нормально распределены - будем использовать специальный метод для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам: scipy.stats.ttest_ind()

Будем использовать общепринятое пороговое значение уровня статистической значимости — 5% 

#### Проверка гипотезы средние пользовательские рейтинги платформ Xbox One и PC одинаковые

- H₀: средние пользовательские рейтинги платформ Xbox One и PC равны  
- H₁: средние пользовательские рейтинги платформ Xbox One и PC не равны  

```python
sample_xone = new_data.loc[(new_data['platform'] == 'XOne') & (~new_data['user_score'].isna()), 'user_score']
sample_pc = new_data.loc[(new_data['platform'] == 'PC') & (~new_data['user_score'].isna()), 'user_score']
print(sample_xone.count())
print(sample_pc.count())
```

Мы выяснили, что наши выборки разного размера поэтому необходимо, указать параметр equal_var = False, чтобы тест не посчитал равными дисперсии выборок. В противном случае тест объединит обе выборки в одну, чтобы поточнее оценить дисперсию.

```python
alpha = .05 # критический уровень статистической значимости


print('Средний пользовательский рейтинг платформы Xbox One', sample_xone.mean())
print('Средний пользовательский рейтинг платформы PC', sample_pc.mean())
results = st.ttest_ind(
    sample_xone, 
    sample_pc, equal_var = False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Уровень p-значения показывает вероятность ~ 11% получить наблюдаемый или еще более далекий от предполагаемого результат при условии, что нулевая гипотеза верна. Таким образом, отвергнуть нулевую гипотезу не удалось.

#### Проверка гипотезы cредние пользовательские рейтинги жанров Action и Sports разные

- H₀: средние пользовательские рейтинги жанров Action и Sports равны  
- H₁: средние пользовательские рейтинги жанров Action и Sports не равны 

```python
sample_action = new_data.loc[(new_data['genre'] == 'Action') & (~new_data['user_score'].isna()), 'user_score']
sample_sports = new_data.loc[(new_data['genre'] == 'Sports') & (~new_data['user_score'].isna()), 'user_score']

print(sample_action.count())
print(sample_sports.count())
```

Мы выяснили, что наши выборки разного размера поэтому необходимо, указать параметр equal_var = False, чтобы тест не посчитал равными дисперсии выборок. В противном случае тест объединит обе выборки в одну, чтобы поточнее оценить дисперсию.

```python
alpha = .05 # критический уровень статистической значимости

print('Средний пользовательский рейтинг жанра Action', sample_action.mean())
print('Средний пользовательский рейтинг жанра Sports', sample_sports.mean())
results = st.ttest_ind(
    sample_action, 
    sample_sports, equal_var = False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Средние пользовательские рейтинги жанров Action и Sports статистически значимо отличаются, вероятность случайно получить такую разницу меньше выбранного уровня значимости, мы отвергаем нулевую гипотезу.
